```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True)
fig.set_figwidth(15)
fig.set_figheight(5)

titles = [r'$x_0$', r'$\dot{x_0}$']
y_vals = [xout[0, :, 0], xout[0, :, 1]]

# axes.flat returns the set of axes as a flat (1D) array instead
# of the two-dimensional version we used earlier
for ax, title, y in zip(axes.flat, titles, y_vals):
    ax.plot(tout[0, :], y)
    ax.set_title(title)
    ax.grid(True)
    ax.set_xlim([0, 10])
    ax.set_ylim([-2, 2])
```

```python
noise_mapping = (0.5, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0005, 0.0001)
for parvalue in range(8):
    with open('./data/noise_' + str(parvalue) + '.pkl', 'rb') as f:
       xout, tout, xout_without_noise, euler_param, sim_param = pickle.load(f)

    x = xout[:, 0::10, :]
    t = tout[:, 0::10]

    fig, axes = plt.subplots(nrows=1, ncols=2, sharex=True)
    fig.set_figwidth(15)
    fig.set_figheight(5)
    titles = [r'$x_0$', r'$\dot{x_0}$']

    for i in range(10):
        y_vals = [xout[i, :, 0], xout[i, :, 1]]
        for ax, title, y in zip(axes.flat, titles, y_vals):
            ax.plot(tout[i, :], y, label='initial condition '+str(euler_param.ic[i]))
            ax.set_title(title)
            ax.grid(True)
            ax.set_xlim([0, 10])
            ax.set_ylim([-2, 2])

    plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
    fig.title('Observed data used for experiments, noise = ' + str(noise_mapping[parvalue]))
    plt.savefig('./data/plots/noise_' + str(parvalue) + '.eps', format = 'eps', bbox_inches='tight')
```
