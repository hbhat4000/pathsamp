```python
import numpy as np
import matplotlib.pyplot as plt

plt.axis([0, 40, 0, 50])

for i in range(10):
    plt.plot(tout[0, :], xout[i, :, 2], label='time series '+str(i))
    plt.pause(0.05)

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, sharex=True)
fig.set_figwidth(15)
fig.set_figheight(5)

titles = [r'$x$', r'$y$', r'$z$']
y_vals = [xout[0, :, 0], xout[0, :, 1], xout[0, :, 2]]

# axes.flat returns the set of axes as a flat (1D) array instead
# of the two-dimensional version we used earlier
for ax, title, y in zip(axes.flat, titles, y_vals):
    ax.plot(tout[0, :], y)
    ax.set_title(title)
    ax.grid(True)
    ax.set_xlim([0, 40])
    ax.set_ylim([-20, 50])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, sharex=True)
fig.set_figwidth(15)
fig.set_figheight(5)
titles = [r'$x$', r'$y$', r'$z$']

for i in range(10):
    y_vals = [xout[i, :, 0], xout[i, :, 1], xout[i, :, 2]]

    # axes.flat returns the set of axes as a flat (1D) array instead
    # of the two-dimensional version we used earlier
    for ax, title, y in zip(axes.flat, titles, y_vals):
        ax.plot(tout[i, :], y, label='time series '+str(i))
        ax.set_title(title)
        ax.grid(True)
        ax.set_xlim([0, 40])
        ax.set_ylim([-20, 50])

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.show()
```
