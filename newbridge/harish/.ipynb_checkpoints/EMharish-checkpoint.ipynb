```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import scipy.special as poly
from math import factorial
```

```python
dof = 4
numsteps = 25000
savesteps = 100
ft = 10.0
ic = np.array([1.0])
it = np.array([0.0])
dt = ft / savesteps

g = 1/2
coef = np.array([1., 2., -1., -0.5])
```

```python
def mypoly(x):
    y = np.zeros((x.size, dof))
    for i in range(x.size):
        for j in range(dof):
            H = poly.hermitenorm(j, monic = True)
            # normalization constant = np.sqrt(np.sqrt(2 * np.pi) * factorial(j))
            # for probabilist's Hermite (hermitenorm)
            # normalization constant = np.sqrt(np.sqrt(np.pi) * (2**j) * factorial(j))
            # for physicists' Hermite (hermite)
            y[i,j] = H(x[i]) / np.sqrt(np.sqrt(2 * np.pi) * factorial(j))
    return y
```

```python
def drift(x, theta):
    return (np.dot(mypoly(x), theta))
```

```python
drift(x=np.array([1.0, 2.0, 3.0]), theta= np.array([1., 2., -1., -0.5]))
```

```python
j = 1
x = np.zeros(savesteps + 1)
t = np.zeros(savesteps + 1)
h = ft / numsteps
h12 = np.sqrt(h)
x[0] = ic
t[0] = it

for i in range(1, numsteps + 1):
    ic = ic + drift(ic, coef) * h + g * h12 * np.random.standard_normal(1)
    if (i % (numsteps / savesteps) == 0):
        x[j] = ic
        t[j] = i*h + it
        j = j + 1
```

```python
print(x)
print(t)
plt.plot(t, x)
```

```python
def brownianbridge(xin, tin, n, index):
    h = (tin[index + 1] - tin[index]) / n
    tvec = tin[index] + (1+np.arange(n))*h
    h12 = np.sqrt(h)
    wincs = np.random.normal(scale=h12, size=n)
    w = np.cumsum(wincs)
    bridge = xin[index] + w - ((tvec - tin[index])/(tin[index + 1]-tin[index]))*(w[n-1] + xin[index] - xin[index + 1])
    tvec = np.concatenate((t[[index]], tvec))
    bridge = np.concatenate((x[[index]],bridge))
    return tvec, bridge
```

```python
# girsanov LOG likelihood
def girsanov(path, dt, theta):
    b = drift(path, theta)
    int1 = np.dot(b[:-1], np.diff(path))
    b2 = np.square(b)
    int2 = np.sum(0.5*(b2[1:] + b2[:-1]))*dt
    r = int1 - 0.5*int2
    return r
```

```python
def bridge1d(x,t,n,i,dt,theta):
    tout, xcur = brownianbridge(x,t,numsubintervals,i)
    lik = girsanov(path=xcur, dt=h, theta=theta)
    return xcur, lik
```

```python
theta = np.array([1., 2., -1., -0.5])
n = t.size
zoomfac = 100
h = (t[1] - t[0]) / zoomfac
bign = (n - 1)*zoomfac + 1
numsubintervals = 100
```

```python
done = False
mytol = 10**(-3)
numiter = 0
numpaths = 10
while (done == False):
    numiter = numiter + 1
    print(numiter)
    mmat = np.zeros((dof, dof))
    rvec = np.zeros(dof)
    for s in range(numpaths):
        samples = np.zeros(bign)
        for i in range(n-1):
            print("i", i, "numiter", numiter, "s", s)
            _, oldlik = bridge1d(x,t,numsubintervals,i,h,theta)
            pathRealized = False
            while(pathRealized == False):
                prop, proplik = bridge1d(x,t,numsubintervals,i,h,theta)
                rho = np.exp(proplik - oldlik)
                if (rho > np.random.uniform()):
                    xcur = prop
                    print(xcur.size)
                    oldlik = proplik
                    pathRealized = True
                    si = i*(n-1)
                    if (i < (n-2)):
                        ei = (i+1)*(n-1)
                        samples[si:ei] = xcur[:(-1)]
                    else:
                        samples[si:bign] = xcur
        
        plt.plot(samples)
        pp = mypoly(samples[:(-1)])
        mmat = mmat + h * np.matmul(pp.T, pp) / numpaths
        rvec = rvec + np.matmul((np.diff(samples)).T, pp) / numpaths

    print(mmat)
    print(rvec)
        
    newtheta = np.linalg.solve(mmat, rvec)
    check = np.sum(np.abs(newtheta - theta))
    if (check < mytol):
        print("finished!")
        print(check)
        print(theta)
        done = True
    theta = newtheta
    print(check)
    print(theta)
```
