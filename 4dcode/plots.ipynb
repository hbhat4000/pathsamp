```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True)
plt.axis([0, 10, -5, 5])
fig.set_figwidth(15)
fig.set_figheight(10)

titles = [r'$\vec{\theta_0}$', r'$\dot{\vec{\theta_0}}$', r'$\vec{\theta_1}$', r'$\dot{\vec{\theta_1}}$']
y_vals = [xout[0, :, 0], xout[0, :, 1], xout[0, :, 2], xout[0, :, 3]]

# axes.flat returns the set of axes as a flat (1D) array instead
# of the two-dimensional version we used earlier
for ax, title, y in zip(axes.flat, titles, y_vals):
    ax.plot(tout[0, :], y)
    ax.set_title(title)
    ax.grid(True)
    ax.set_xlim([0, 10])
    ax.set_ylim([-5, 5])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True)
fig.set_figwidth(15)
fig.set_figheight(10)
titles = [r'$\vec{\theta_0}$', r'$\dot{\vec{\theta_0}}$', r'$\vec{\theta_1}$', r'$\dot{\vec{\theta_1}}$']

for i in range(10):
    y_vals = [xout[i, :, 0], xout[i, :, 1], xout[i, :, 2], xout[i, :, 3]]

    # axes.flat returns the set of axes as a flat (1D) array instead
    # of the two-dimensional version we used earlier
    for ax, title, y in zip(axes.flat, titles, y_vals):
        ax.plot(tout[i, :], y, label='time series '+str(i))
        ax.set_title(title)
        ax.grid(True)
        ax.set_xlim([0, 10])
        ax.set_ylim([-10, 10])

plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.show()
```
